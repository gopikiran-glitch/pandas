```python
df_copy['size'] = df_copy['size'].replace('Varies with device', np.nan)

# Remove 'M' and 'k' letters, handle both upper/lower cases
df_copy['size'] = df_copy['size'].str.replace('M', '', regex=False)
df_copy['size'] = df_copy['size'].str.replace('k', '', regex=False)
df_copy['size'] = df_copy['size'].str.replace('G', '', regex=False)

# Convert column to numeric (automatically makes invalid entries NaN)
df_copy['size'] = pd.to_numeric(df_copy['size'], errors='coerce')
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Price'].unique()
```

```python
df_copy['Installs']=df_copy['Installs'].str.replace('+','')
df_copy['Price']=df_copy['Price'].str.replace('$','')
df_copy['Installs']=df_copy['Installs'].str.replace(',','')
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Last Updated'].unique()
```

```python
df_copy['Last Updated']=pd.to_datetime(df_copy['Last Updated'])
```

```python
df_copy.to_csv('cleanplaystoredataset.csv')
```

```python
df_copy=df_copy.drop_duplicates(subset=['App'],keep='first')
```

```python
df_copy.shape
```

```python
numeric_features = [feature for feature in df_copy.columns if df_copy[feature].dtype != 'O']
categorical_features = [feature for feature in df_copy.columns if df_copy[feature].dtype == 'O']


print("We have {} numerical features".format(len(numeric_features)))
print("We have {} categorical features".format(len(categorical_features)))
```

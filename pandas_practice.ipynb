```python
import pandas as pd
data = {
    'Department': ['HR', 'HR', 'IT', 'IT', 'Finance', 'Finance', 'Finance'],
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace'],
    'Salary': [50000, 55000, 60000, 65000, 70000, 72000, 71000],
    'Experience': [3, 4, 5, 7, 10, 9, 8]
}

df = pd.DataFrame(data)
print(" Original DataFrame:")
print(df)
grouped = df.groupby('Department')
print("\n Grouped object:")
print(grouped)
```

```python

print("\n Average Salary by Department:")
print(df.groupby('Department')['Salary'].mean())

print("\n Employee Count by Department:")
print(df.groupby('Department')['Employee'].count())

print("\nTotal Experience by Department:")
print(df.groupby('Department')['Experience'].sum())
```

```python

print("\n Multiple Aggregations on Salary:")
print(df.groupby('Department')['Salary'].agg(['min', 'max', 'mean', 'count']))
print("\nAggregating Multiple Columns:")
```

```python

employees = pd.DataFrame({
    'EmpID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'DeptID': [101, 102, 102, 103]
})

departments = pd.DataFrame({
    'DeptID': [101, 102, 103, 104],
    'DeptName': ['HR', 'IT', 'Finance', 'Admin']
})

print(" Employees:")
print(employees)
print("\nDepartments:")
print(departments)
merged_inner = pd.merge(employees, departments, on='DeptID', how='inner')
print("\n Inner Join Result:")
print(merged_inner)
```
